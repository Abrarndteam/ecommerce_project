```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_excel('/Online Retail.xlsx')
print(df)
df.isnull().sum()
df.dropna(inplace=True)
df.duplicated().sum()
df.drop_duplicates(inplace=True)
df['UnitPrice'] = df['UnitPrice'].apply(lambda x: max(x, 0))
```

```python
# prompt: print summary of data

print(df.info())
```

```python
# prompt: find centeral tendency and dispersion

# Central tendency
mean = df['UnitPrice'].mean()
median = df['UnitPrice'].median()
mode = df['UnitPrice'].mode()[0]

# Dispersion
range_ = df['UnitPrice'].max() - df['UnitPrice'].min()
variance = df['UnitPrice'].var()
std_dev = df['UnitPrice'].std()

# Print the results
print("Central tendency:")
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print("\nDispersion:")
print(f"Range: {range_}")
print(f"Variance: {variance}")
print(f"Standard deviation: {std_dev}")
```

```python
# prompt: find despersion,quantity and unit price

# Calculate dispersion
range_ = df['UnitPrice'].max() - df['UnitPrice'].min()
variance = df['UnitPrice'].var()
std_dev = df['UnitPrice'].std()

# Calculate quantity
quantity = df['Quantity'].sum()

# Calculate unit price
unit_price = df['UnitPrice'].mean()

# Print the results
print("Dispersion:")
print(f"Range: {range_}")
print(f"Variance: {variance}")
print(f"Standard deviation: {std_dev}")
print("\nQuantity:")
print(f"Total quantity: {quantity}")
print("\nUnit Price:")
print(f"Average unit price: {unit_price}")
```

```python
# prompt: code to make histogram charts of quantity and its frequency

import matplotlib.pyplot as plt

# Create a new column for frequency
df['Quantity_Frequency'] = df.groupby('Quantity')['Quantity'].transform('count')

# Plot the histogram
df['Quantity'].hist(bins=10, edgecolor='black')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.title('Histogram of Quantity')
plt.show()
```

```python
# prompt: code for graph of unit price and frequency

import matplotlib.pyplot as plt
# Create a new column for frequency
df['UnitPrice_Frequency'] = df.groupby('UnitPrice')['UnitPrice'].transform('count')

# Plot the histogram
df['UnitPrice'].hist(bins=10, edgecolor='black')
plt.xlabel('Unit Price')
plt.ylabel('Frequency')
plt.title('Histogram of Unit Price')
plt.show()
```

```python
# prompt: code to draw a chart to distribution of no of transactions countrywise

df_country = df.groupby('Country')['InvoiceNo'].count().reset_index()
df_country.columns = ['Country', 'No_of_Transactions']

fig = px.bar(df_country, x='Country', y='No_of_Transactions',
             title='Distribution of Number of Transactions by Country',
             color_discrete_sequence=px.colors.qualitative.Pastel1)
fig.update_layout(
    title_font=dict(size=16),
    xaxis_title_font=dict(size=14),
    yaxis_title_font=dict(size=14),
    xaxis_tickangle=-45,
    yaxis=dict(tickfont=dict(size=12)),
    margin=dict(l=80, r=40, t=80, b=30),
    plot_bgcolor='white',
    paper_bgcolor='white',
    showlegend=False
)
fig.update_xaxes(tickfont=dict(size=10), showline=True, linewidth=1, linecolor='black')
fig.update_yaxes(showgrid=True, gridwidth=0.5, gridcolor='gray')

fig.show()
```

```python
# prompt: draw a graph for quantity vs unit price scatterplot

import matplotlib.pyplot as plt
sns.scatterplot(x="Quantity", y="UnitPrice", data=df)
plt.xlabel("Quantity")
plt.ylabel("Unit Price")
plt.title("Quantity vs Unit Price")
plt.show()
```

```python
# prompt: identify the busiest months and days of the weeks in terms of sales

import pandas as pd
# Extract month and day of the week from the InvoiceDate column
df['Month'] = pd.to_datetime(df['InvoiceDate']).dt.month_name()
df['DayOfTheWeek'] = pd.to_datetime(df['InvoiceDate']).dt.day_name()

# Group by month and calculate total quantity sold
monthly_sales = df.groupby('Month')['Quantity'].sum().reset_index()

# Sort by total quantity sold and get the top 3 busiest months
busiest_months = monthly_sales.sort_values(by='Quantity', ascending=False).head(3)

# Print the busiest months
print("Busiest Months:")
for month in busiest_months['Month']:
    print(f"\t- {month}")

# Group by day of the week and calculate total quantity sold
daily_sales = df.groupby('DayOfTheWeek')['Quantity'].sum().reset_index()

# Sort by total quantity sold and get the top 3 busiest days
busiest_days = daily_sales.sort_values(by='Quantity', ascending=False).head(3)

# Print the busiest days
print("\nBusiest Days of the Week:")
for day in busiest_days['DayOfTheWeek']:
    print(f"\t- {day}")
```

```python
# prompt: explore the top selling products and countries based on quantity sold with quantity

# Top selling products by quantity
top_selling_products = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)

print("Top Selling Products by Quantity:")
for product, quantity in top_selling_products.items():
    print(f"\t- {product}: {quantity}")

# Top selling countries by quantity
top_selling_countries = df.groupby('Country')['Quantity'].sum().sort_values(ascending=False).head(10)

print("\nTop Selling Countries by Quantity:")
for country, quantity in top_selling_countries.items():
    print(f"\t- {country}: {quantity}")
```

```python
# prompt: identify any outliers or anomalies in the dataset and discuss their potential impact on analysis

import numpy as np

# Calculate the z-score for each data point
z_scores = np.abs((df['UnitPrice'] - df['UnitPrice'].mean()) / df['UnitPrice'].std())

# Identify outliers based on a threshold
outliers = df[z_scores > 3]

# Print the number of outliers
print(f"Number of outliers: {len(outliers)}")

# Print the outlier data points
print(outliers)

# Potential impact of outliers on analysis
# - Outliers can distort the mean and standard deviation of the data
# - Outliers can lead to incorrect conclusions about the distribution of the data
# - Outliers can affect the results of statistical tests
# - Outliers can make it difficult to identify patterns and trends in the data

# Dealing with outliers
# - Outliers can be removed from the dataset
# - Outliers can be replaced with the mean or median of the data
# - Outliers can be transformed using a logarithmic or square root transformation
```
